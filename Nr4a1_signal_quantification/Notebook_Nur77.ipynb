```python
metrics = threshold_values.copy()
metrics.insert(2,"pixels_of_signal",np.zeros(40))
metrics.insert(2,"perc_signal_in_all_vol",np.zeros(40))
metrics.insert(2,"raw_int_den",np.zeros(40))
metrics.insert(2,"mean",np.zeros(40))
metrics.insert(2,"intden",np.zeros(40))

metrics.insert(2,"raw_int_den_pixels",np.zeros(40))
metrics.insert(2,"mean_div_pixels",np.zeros(40))
metrics.insert(2,"intden_div_thr_pixels",np.zeros(40))

metrics.insert(2,"raw_int_den_thresholded",np.zeros(40))
metrics.insert(2,"mean_div_thresholded",np.zeros(40))
metrics.insert(2,"intden_div_thr_thresholded",np.zeros(40))

metrics.insert(2,"pixels_of_signal_per_slice",np.zeros(40))
metrics.insert(2,"mean_C02_signal_value",np.zeros(40))
metrics.insert(2,"mean_C02_signal_value_in_ROI",np.zeros(40))


metrics
```

```python
#compute mean sig/PI and sig/Autofluorescence in general and in thresholded areas, maybe of interest

#sig/sf:
folder_C02 = '/media/izabela/18TB_linux1/nur77 12x scans/Nifti_files/C02/'
folder_af = '/media/izabela/18TB_linux1/nur77 12x scans/Nifti_files/C00/'
fodler_C01 = '/media/izabela/18TB_linux1/nur77 12x scans/Nifti_files/vols_C01/'

folder_threshold_masks = '/media/izabela/18TB_linux1/nur77 12x scans/Nifti_files/threshold_masks/'

folder_thresholded_vols = '/media/izabela/18TB_linux1/nur77 12x scans/Nifti_files/thresholded_vols/'

folder_foreground_masks = '/media/izabela/18TB_linux1/nur77 12x scans/Nifti_files/foreground_masks/'

for index, filename in enumerate(threshold_values[0]):
    if filename in scans_of_interest:
        thresh = threshold_values[1][index]
        
        if(thresh):
            one_vol_C02=filehandling.readNifti(folder_C02+filename)
            
            binary_mask = (one_vol_C02>thresh).astype(np.uint8)
            
            one_vol_C01=filehandling.readNifti(fodler_C01+filename)
            one_vol_C00=filehandling.readNifti(folder_af+filename)
            
            
            metrics["pixels_of_signal"][index] = np.sum(binary_mask)
            metrics["perc_signal_in_all_vol"][index] = np.sum(binary_mask)/np.prod(binary_mask.shape)*100
            
            metrics["raw_int_den"][index]= np.sum(one_vol_C02)
            metrics["mean"][index]=  np.sum(one_vol_C02)/np.prod(one_vol_C02.shape)
            metrics["intden"][index]= np.sum(one_vol_C02)/np.prod(one_vol_C02.shape)*(1.84*1.84*5)
            
            metrics["raw_int_den_pixels"][index]= np.sum(binary_mask)
            metrics["mean_div_pixels"][index]=  np.sum(binary_mask)/np.prod(binary_mask.shape)
            metrics["intden_div_thr_pixels"][index]= np.sum(binary_mask)/np.prod(binary_mask.shape)*(1.84*1.84*5)
            
            
            metrics["raw_int_den_thresholded"][index]= np.sum(one_vol_C02[one_vol_C02>thresh])
            metrics["mean_div_thresholded"][index]=  np.sum(one_vol_C02[one_vol_C02>thresh])/np.prod(one_vol_C02[one_vol_C02>thresh].shape)
            metrics["intden_div_thr_thresholded"][index]= np.sum(one_vol_C02[one_vol_C02>thresh])/np.prod(one_vol_C02[one_vol_C02>thresh].shape)*(1.84*1.84*5)
            
            metrics["pixels_of_signal_per_slice"][index]= np.sum(binary_mask)/binary_mask.shape[2]
            metrics["mean_C02_signal_value"][index]=  np.mean(one_vol_C02)
            metrics["mean_C02_signal_value_in_ROI"][index]= np.mean(one_vol_C02[one_vol_C02>thresh])
            

            
        
```
